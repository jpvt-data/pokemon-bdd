```python
# ============================================================
# [00_MASTER] — ORCHESTRATEUR PROJET POKEMON
# Lance tous les notebooks dans l'ordre
# Prismora Solutions — 2026
# ============================================================

import papermill as pm
from pathlib import Path
from datetime import datetime

ROOT = Path.cwd()
LOG_DIR = ROOT / "data/logs"
LOG_DIR.mkdir(exist_ok=True)

# Liste des notebooks à exécuter dans l'ordre
NOTEBOOKS = [
    "01_SETUP_BDD.ipynb",
    "02_POCKET_CARDS.ipynb",
    # "03_TCG_CARDS.ipynb",     # décommenter quand prêt
    # "04_POKEDEX.ipynb",       # décommenter quand prêt
    # "05_IMAGES.ipynb",        # décommenter quand prêt
]

print("=" * 55)
print("  PROJET POKEMON — Exécution complète")
print(f"  {datetime.now().strftime('%d/%m/%Y %H:%M:%S')}")
print("=" * 55)

for nb in NOTEBOOKS:
    nb_path = ROOT / nb
    if not nb_path.exists():
        print(f"\n⏭️  {nb} → IGNORÉ (fichier absent)")
        continue
    print(f"\n▶️  Lancement : {nb} ...")
    try:
        pm.execute_notebook(
            str(nb_path),
            str(LOG_DIR / f"output_{nb}"),  # sauvegarde log d'exécution
            kernel_name="python3"
        )
        print(f"✅ {nb} → OK")
    except Exception as e:
        print(f"❌ {nb} → ERREUR : {e}")
        break  # stoppe si un notebook plante

print("\n" + "=" * 55)
print("  Exécution terminée")
print("=" * 55)
```
